# Brief Explanation about 🤗 Transformers library

**Huggingface is the place -- for open-source generative AI models.**

 Hugging Face is a machine learning and data science platform empowers users to build, deploy, and train machine learning models. Learn how developers continuously expand its repository, from language models to computer vision and audio models, and benefit from an active open-source community.

This powerful tool downloads pretrained models tailored for tasks like Natural Language Understanding (NLU), including sentiment analysis, and Natural Language Generation (NLG), which involves tasks like text completion and language translation.

Initially, we'll demonstrate the simplicity of utilizing the pipeline API to swiftly apply these pretrained models for inference. Following that, we'll delve deeper into how the library provides access to these models and facilitates data preprocessing for your specific needs.

The easiest way to use a pretrained model on a given task is to use pipeline. 🤗 Pre-trained model based on Transformers application -

Sentiment analysis: is a text positive or negative?

```python
from transformers import pipeline
classifier = pipeline('sentiment-analysis')
```

The pipelines are a great and easy way to use models for inference.

# Pre-trained model

default: "distilbert-base-uncased-finetuned-sst-2-english"  

Link: https://huggingface.co/distilbert/distilbert-base-uncased-finetuned-sst-2-english

another model:"bert-base-multilingual-uncased-sentiment"

Multilingual: English, Dutch, German, French, Italian






```python
from transformers import pipeline
classifier = pipeline('sentiment-analysis', model = "nlptown/bert-base-multilingual-uncased-sentiment")
```

```python
text1 = "We are very happy to show you the implementation of 🤗 Transformers library"
classifier(text1)
```

```python
text1_sp = "Estamos muy contentos de mostrarles la implementación de la biblioteca Transformers."
classifier(text1_sp)
```

```python
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification
```

```python
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"
model = TFAutoModelForSequenceClassification.from_pretrained(model_name, from_pt = True)
tokenizer = AutoTokenizer.from_pretrained(model_name)
classifier = pipeline('sentiment-analysis', model = model, tokenizer =tokenizer )
```
